### FIFA World Cup qualifications 1994-2022

```python



web = 'https://en.wikipedia.org/wiki/1994_FIFA_World_Cup_qualification_%E2%80%93_UEFA_Group_6#Sweden_v_Israel'
response = requests.get(web)
content = response.text
soup = BeautifulSoup(content, 'lxml')
    
matches = soup.find_all('div', class_='footballbox')
    
home = []
score = []
away = []
    
for match in matches:
    home.append(match.find('th', class_='fhome').get_text())
    score.append(match.find('th', class_='fscore').get_text())
    away.append(match.find('th', class_='faway').get_text())
    
dict_football = {'home': home, 'score': score, 'away': away}
df_football_1994 = pd.DataFrame(dict_football)
df_football_1994['years'] = '1994'   
df_football_1994
```

```python
# create a condition for row selection
israel_condition = (all_matches['home'] == 'Israel\xa0') | (all_matches['away'] == '\xa0Israel')

# apply the condition using the loc method
israel_matches = all_matches.loc[israel_condition]
israel_matches
```

```python
israel_matches.to_csv('israel_matches.csv', index=False)
```
